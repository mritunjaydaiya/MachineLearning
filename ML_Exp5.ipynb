```python
import matplotlib.pyplot as plt
```

```python
x=[2,9,5,5,3,7,1,8,6,2]
y=[69,98,82,77,71,84,55,94,84,64]
```

```python
def mean(l):
  return sum(l)/len(l)
```

```python
def variance(l):
  l_= mean(l)
  return sum(list(map(lambda n : (n-l_)**2 ,l))) 
```

```python
def covariance(x,y):
  #cv=[]
  cv=0
  for i in range(len(x)):
    #cv.append((x[i] - mean(x))*(y[i] - mean(y)))
    cv+=(x[i] - mean(x))*(y[i] - mean(y))
  return cv
```

```python
m = covariance(x,y)/variance(x)
c = mean(y) - m*mean(x)
```

```python
yp=[]
for i in range(len(x)):
  yp.append(m*x[i] + c)
```

```python
ssr= sum(map(lambda n : (n-mean(y))**2,yp))
sst= variance(y)

r2= ssr/sst
```

```python
print('Accuracy is :',r2)
```

```python
plt.scatter(x,y)
plt.plot(x, yp)
plt.show()
```
