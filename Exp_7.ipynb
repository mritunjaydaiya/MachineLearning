```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("EXP7.csv")
data
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['WC']= label_encoder.fit_transform(data['WC'])
data['HU']= label_encoder.fit_transform(data['HU'])
data['WT']= label_encoder.fit_transform(data['WT'])
data['WC'].unique()
data['HU'].unique()
data['WT'].unique()
data
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
y_pred
```

```python
from sklearn import  metrics
print("Naive Bayesian accuracy is :",metrics.accuracy_score(y_test,y_pred)*100)
```
